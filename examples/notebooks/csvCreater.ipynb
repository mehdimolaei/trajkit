```python
csvfolder = "/Users/mehdi/air-water/data/csv/experiment_001_2017-08-16/"
datasetID = "exp001_t043m_r01um_2017-08-16"
csvfile = f"{csvfolder}{datasetID}.csv.gz"
df = pd.read_csv(csvfile)
trajset = TrajectorySet.from_dataframe(
    df,
    dataset_id=datasetID,
    track_id_col="id",
    position_cols=["x", "y"],
    time_col="t",
    frame_col="t",
    units={"t": "frame", "x": "pixel"},
)
output_dir = Path(f"/Users/mehdi/air-water/data/parquet/experiment_001_2017-08-16/{datasetID}")
output_dir.mkdir(parents=True, exist_ok=True)

# Assume you already have `trajset` in memory
save_trajectory_set(trajset, output_dir)

# print(f"Saved trajectory set to {output_dir}")
```
